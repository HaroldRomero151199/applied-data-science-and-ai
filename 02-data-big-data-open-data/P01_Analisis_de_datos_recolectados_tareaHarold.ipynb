Generamos la primera estadística

```python
# Calcular la edad promedio
edad_promedio = datos['Edad'].mean()
edad_min = datos['Edad'].min()
edad_max = datos['Edad'].max()
print("Edad promedio:", edad_promedio)
print("Edad mínima:", edad_min)
print("Edad mínima:", edad_max)
```

```python
# Ahora con numpy generamos otras estadísticas

# Seleccionamos el nombre de la columna, en este caso será la altura de las personas
data_array = datos['Altura'].dropna().to_numpy()
# data_array = datos['Altura'].to_numpy()

print(data_array)
mean_value = np.mean(data_array)
median_value = np.median(data_array)
std_value = np.std(data_array)
min = np.min(data_array)
max = np.max(data_array)
q1 = np.quantile(data_array,q=0.25)
q3 = np.quantile(data_array,q=0.75)
print("A partir del dato de altura de las personas, se obtienen las siguientes estadísticas:")
print(f"Media: {mean_value}, Mediana: {median_value}, Desviación Estándar: {std_value}, Mínimo {min}, Máxima {max}, Q1 {q1},Q3 {q3} ")

```

Generamos un histograma a partir del campo Edad y desplegamos el gráfico

```python
# Histograma de la edad
plt.hist(datos['Edad'])
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.title('Distribución de la edad')
plt.show()
```

```python
# Diagrama de barras para la distribución de frecuencias por conocimiento de Big Data

data_array = datos['NivelC_BigData'].dropna().to_numpy()

plt.hist(data_array, bins=10, alpha=0.5)
plt.title('Histograma')
plt.xlabel('NivelC_BigData')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Diagrama de barras para la distribución de frecuencias por conocimiento de SIG

data_array = datos['NivelC_SIG'].dropna().to_numpy()

plt.hist(data_array, bins=10, alpha=0.5)
plt.title('Histograma')
plt.xlabel('NivelC_SIG')
plt.ylabel('Frecuencia')
plt.show()
```

```python
## Alternativa de librería para interacción de datos.

import plotly.express as px

# Here we use a column with categorical data
fig = px.histogram(datos, x="NivelC_BigData")
fig.show()
```

```python
fig = px.histogram(datos, x="NivelC_SIG")
fig.update_layout(bargap=0.2)
fig.show()
```

## 📊 Práctica
Generar una gráficas de histograma de frecuecia con los campos "Entorno_programacion"

```python
data_arrayEP = datos['Entorno_programacion'].dropna().to_numpy()
plt.figure(figsize=(30, 6))
plt.hist(data_arrayEP, bins=10, alpha=0.5)
plt.title('Histograma')
plt.xlabel('Entorno_programacion')
plt.ylabel('Frecuencia')
plt.show()
```

Calcular estadísticas a partir del campo "Peso" y calcula "promedio", "mediana", "min", "max"

```python
data_arrayPK = datos['Peso_kg'].dropna().to_numpy()

print(data_arrayPK)
mean_value = np.mean(data_arrayPK)
median_value = np.median(data_arrayPK)
std_value = np.std(data_arrayPK)
min = np.min(data_arrayPK)
max = np.max(data_arrayPK)
print("A partir del dato de PESO KG de las personas, se obtienen las siguientes estadísticas:")
print(f"Media: {mean_value}, Mediana: {median_value}, Desviación Estándar: {std_value}, Mínimo {min}, Máxima {max}")
```

## 📝Extra 1
Realizar un diagrama de cajas a partir del dato "Peso"

```python
data_array = datos['Peso_kg'].dropna().to_numpy()
plt.boxplot(data_array, vert=False, patch_artist=True)
plt.title('Histograma')
plt.xlabel('NivelC_Peso')
plt.ylabel('Frecuencia')
plt.show()
```

## 📝Extra 2
Realizar un diagrama de dispersión con el campo "Edad" y "Peso" y obtnén una análisis del resultado

```python
age_arrayX = datos['Edad'].dropna().to_numpy()
peso_arrayY = datos['Peso_kg'].dropna().to_numpy()

plt.scatter(age_arrayX, peso_arrayY,  c='red')
plt.title('Diagrama de Dispersión de Edad vs Peso')
plt.xlabel('Edad')
plt.ylabel('Peso Kg')
plt.grid(True)
plt.show()
```
